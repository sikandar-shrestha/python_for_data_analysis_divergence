```python
'''
Task 9: Display the data after it has been sorted by the Stars column 
from high values to low values
'''

df.sort_values('Stars',ascending=False)
```

```python
# Task 10: Replace "USA" with "United States" in the Country column
df['Country'].replace('USA', 'United States', inplace=True)
df.head()
```

```python
# questions
# 1. How many countries are represented in the data?
df['Country'].nunique()
```

```python
# 2. Which three countries have the highest average rating?
df.groupby('Country')['Stars'].mean().nlargest(3)
```

```python
# 3. Which three countries have the lowest average rating?
df.groupby('Country')['Stars'].mean().nsmallest(3)
```

```python
'''
4. Which three countries have the most brands, and how many brands
does each of these countries have?
'''

df.groupby('Country')['Brand'].nunique().nlargest(3)
```
