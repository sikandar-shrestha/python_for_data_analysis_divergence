```python
# https://www.kaggle.com/residentmario/ramen-ratings
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import seaborn as sns
```

## Tasks

```python
# 1 Read the data from the CSV file into a DataFrame and display the first five rows.
data = pd.read_csv('ramen-ratings.csv')
data.head()
```

```python
# 2 Group the data and get the average of the groups to produce a DataFrame that looks like this:
data.groupby(['Country','Brand','Style'])[['Stars']].mean()
```

```python
# 3 Group the data in the same way, but this time get both the average and the count.
data.groupby(['Country','Brand','Style'])[['Stars']].agg(['mean','count'])
```

```python
# 4 Create a pivot table that shows the average stars by ramen style for each country.
data.pivot_table(index='Country', columns='Style', values='Stars', aggfunc='mean')
```

```python
# 5: Calculate the total number of stars for each country, 
# and then rank the countries in descending order by total stars.
data_grouped = data.groupby('Country').sum()
data_grouped.Stars.rank(ascending=False)
```

```python
# 6:Bin the stars data into 11 bins. Label the bins from 0.0 to 5.0 in increments of .5, 
# and then display the count of the stars in each bin.
pd.cut(data.Stars, bins=11, labels=[0,.5,1,1.5,2,2.5,3,3.5,4,4.5,5]).value_counts()
```

```python
# 7: Bin the stars data into five quantiles. Give each bin a descriptive label, 
# and then display the count of the stars in each bin.
pd.qcut(data.Stars, q=5, labels=['Terrible','Bad','Mediocre','Good','Excellent']).value_counts()
```
