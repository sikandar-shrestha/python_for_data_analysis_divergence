### # Task 8: Multiply the Total Volume and AveragePrice columns, and store the result in a new column called EstimatedRevenue. Then, display the first five rows of this data to confirm that the column was added and has the correct values


```python
df.head(3)
```

```python
df['EstimatedRevenue'] = df['Total Volume'] * df['AveragePrice']
```

###  Task 9: Create a DataFrame thatâ€™s grouped by region and type and that includes the average price for the grouped columns. Then, reset the index and display the first five rows


```python
df.groupby(['region', 'type']).agg({'AveragePrice': 'mean'}).reset_index().head()
```

### Task 10: Create a bar plot that shows the mean, median, and standard deviation of the Total Volume column by year
 

```python
df.head()
```

```python
# using pandas visualization method()
volume_stats = df.groupby('year')['Total Volume'].agg(['mean', 'median', 'std'])
volume_stats.plot(kind='bar')
```

### Q2. What is the average price for each type of avocado (organic and conventional)?
 

```python
df.groupby('type')['AveragePrice'].mean()
```

### Q3. Which region has the lowest average price for organic avocados?


```python
df[df['type'] == 'organic'].nsmallest(1, 'AveragePrice')

```

### Q4. Have the Total Bags sold per year of each type of avocado become more or less consistent over time?


```python
df.groupby(['year', 'type'])['Total Bags'].sum().unstack().plot(kind='line', marker='o')
```
