```python
#remove pink warning box
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

## Tasks

### Q1)Read the data from the CSV file into a DataFrame.

```python
data = pd.read_csv('gold.csv')
```

### Q2)Run the info() method, and note the data types of the two columns.

```python
data.info()
```

```python
data.head(5)
```

### Q3) Convert the Date column to the DateTime data type. Then, display the first five rows of data, and note that the dates are for the first of each month starting in 1950.

```python
data.Date = pd.to_datetime(data.Date)
data.head()
```

### Q4)Plot the data with a Pandas line plot.

```python
data.plot(x='Date', y='Price')
```

### Q5) Index the data on the Date column. Then, generate time periods for the first day of each year from 1950 to 2020.

```python
data.set_index('Date', inplace=True)
years = pd.date_range('1950-01-01', '2020-01-01', freq='YS')
```

```python
data.head(3)
```

### Q6)Reindex the data using the new time periods, and display the first five rows again to see that there is just one row for each year.

```python
data.reindex(years).head()
```

### Q7)Plot the reindexed data with a Pandas line plot.

```python
data.reindex(years).plot()
```

### Q8)Resample the data to a quarterly frequency and plot the data again.

```python
data.resample(rule='Q').mean().plot()
```

### Q9)Resample the data to a yearly frequency and plot the data one more time.

```python
data.resample(rule='Y').mean().plot()
```

### Q10)Plot the rolling mean for the data using a Pandas line plot. Experiment with different values for the window parameter.

```python
# 10 part 1
data.rolling(window=6).mean().plot()
```

```python
# 10 part 2
data.rolling(window=12).mean().plot()
```

```python
# 10 part 3
data.rolling(window=30).mean().plot()
```
