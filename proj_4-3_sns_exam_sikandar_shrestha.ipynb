```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import seaborn as sns
```

```python
#Task1:
# Read the data from the CSV file into a DataFrame and display the first five rows.
df=pd.read_csv('exams.csv')
df.head()
```

```python
df.columns.unique()
```

```python
#Task2:
# Use the Seaborn catplot() method to create a plot like the one below. 
# To make that easy to do, you can use a count plot.
g=sns.catplot(data=df,x='parental level of education',kind='count',hue='parental level of education')
```

```python
#Task3:
# Rotate the x labels for the above plot to make them readable.
# Rotate x-axis labels
g=sns.catplot(data=df,x='parental level of education',kind='count',hue='parental level of education')
for ax in g.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
#Task4:
# Create a scatter plot with a specific method that compares the writing score with the reading score.
# Use a different color to show which students took a test prep course.
sns.scatterplot(data=df, x='reading score', y='writing score', hue='test preparation course')

            
```

```python
#Task5:Create the same plot with a general method.
sns.relplot(data=df, x='reading score', y='writing score', hue='test preparation course')
```
