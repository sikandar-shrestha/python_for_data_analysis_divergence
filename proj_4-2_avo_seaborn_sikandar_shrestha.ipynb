```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import seaborn as sns
```

```python
#Q1: Read the data from the CSV file into a DataFrame and display the first five rows.
df=pd.read_csv('avocado.csv')
df.head()
```

### Q2:Create the following plot with a Seaborn specific method.
(image is not showing)


```python
g=sns.scatterplot(x='AveragePrice', y='Total Volume', data=df)
g.set(title="scatter plotting between 'average price' vs 'total volume' ",
      ylabel='total volume',xlabel='average price')
```

```python
#Q3:Create the same plot with a generic method.
g=sns.relplot(data=df,kind='scatter',
              x='AveragePrice',
              y='Total Volume',
             )
```

```python
#Q4:Create the same plot but with the hue parameter set to year and 
# the dots for the total volume ranging in size from 10 to 100.
g=sns.relplot(data=df,kind='scatter',
              x='AveragePrice',
              y='Total Volume',
              hue='year'
             )
```
