```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
```

```python
#Task1: Read the data from the CSV file into a DataFrame and display the first five rows.

df=pd.read_csv('exams.csv')
df.head(5)
```

```python
#Task2:Display the basic information for the DataFrame and 
#      its columns using the info() method.
df.info()
```

```python
#Task3:Display statistical information for the math score, reading score, 
# and writing score columns using the describe() method.

df.describe()
```

```python
#Task4:Group the data by the race/ethnicity column and display the mean scores.
df.groupby('race/ethnicity')[['math score','reading score','writing score']].mean()
```

```python
# another way
df[['race/ethnicity','math score','reading score',
    'writing score']].groupby('race/ethnicity').mean()
```

```python
#Task5:Display a single column as a DataFrame with bracket notation.
df[['gender']]
```

```python
#Task6:Display a single column as a Series with bracket notation.
df['gender']
```

```python
#Task7:Display a single column as a Series with dot notation.
df.gender
```

```python
#Task8:Display only rows for females with a math score greater than or equal to 90.

df[(df['gender'] == 'female') & (df['math score'] >= 90)]
```

## Questions

## Q1: Does taking a test preparation course improve average scores?

```python
df[['test preparation course','math score','reading score',
    'writing score']].groupby('test preparation course').mean()
```

## Q2: Which gender is better on average at math?

```python
df.groupby('gender')['math score'].mean()
```

```python
df.groupby('gender')['math score'].mean()
```

## Q3: Which gender is better on average at all three subjects? 

Hint: Start by adding a column to the DataFrame with the total score.

```python
df['total score'] = df['math score'] + df['reading score'] 
+ df['writing score']
```

```python
df.head(3)
```

```python
df.groupby('gender')['total score'].mean()
```

## Q4: Does the parentsâ€™ level of education have an effect on the average scores?

```python
df[['parental level of education','math score','reading score',
    'writing score']].groupby('parental level of education').mean()
```
