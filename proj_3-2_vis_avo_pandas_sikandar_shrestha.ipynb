```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
```

## Tasks

```python
#Task1:Read the data from the .csv file into a DataFrame and display the first five rows.
data = pd.read_csv('avocado.csv')
data.head()
```

```python
#Task2:Create a new DataFrame that contains the total of the Small Bags, Large Bags, 
#and XLarge Bags columns grouped by type, and then display the DataFrame.
dataGrouped = data.groupby('type').sum()[['Small Bags','Large Bags','XLarge Bags']]
dataGrouped
```

```python
#Task3:Use the grouped data to create a bar plot that shows the number of small, large, and extra-large bags for both types of avocado.
dataGrouped.plot.bar()
```

```python
# another way
data.groupby('type').sum()[['Small Bags','Large Bags','XLarge Bags']].plot.bar()
```

```python
# another way
data.groupby('type')[['Small Bags','Large Bags','XLarge Bags']].sum().plot.bar()
```

```python
#Task4:Use the original data to create a scatter plot for the Total Volume and AveragePrice columns.
data.plot.scatter(x='Total Volume', y='AveragePrice')
```

## Question

```python
#Q1:In what year were the most bags of avocados sold?
data.groupby('year')['Total Bags'].sum().plot.bar()
```

```python
#Q2:Do the total bags of avocados appear to have an effect on the average price?
data.plot.scatter(x='Total Bags', y='AveragePrice')
```
